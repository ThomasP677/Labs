```python
import random

def mid_point(string):
    return len(string)/2




def count_character(s, c):
    counter = 0
    for i in range(len(s)):
        if s[i] == c:
            counter += 1
    return counter

def divid_strings(string):
    for i in range(len(string)):
        if string[i] == ' ':
            new_string = string[0:i]
            return new_string

def divid_strings2(string):
    for i in range(len(string)):
        if string[i] == ' ':
            new_string = string[i:len(string)]
            return new_string

def repeat_strings(s,n):
    out = s * n
    return out

def new_name1(name1,name2):
    new_string = divid_strings(name1)
    new_string2 = divid_strings(name2)
    string_int = int(mid_point(new_string))
    string_int2 = int(mid_point(new_string2))
    final_string = new_string[0:string_int]
    final_string2 = new_string2[string_int2: len(new_string2)]
    name = final_string + final_string2
    return name

def new_last_name(name1,name2):
    new_string = divid_strings2(name1)
    new_string2 = divid_strings2(name2)
    string_int = int(mid_point(new_string))
    string_int2 = int(mid_point(new_string2))
    final_string = new_string[0:string_int + 1]
    final_string2 = new_string2[string_int2: len(new_string2)]
    name = final_string + final_string2
    return name

# Debugging function calls. You can add more or change
# The following should return 3
char_count = count_character('banana', 'a')
print(char_count)

# The following call should return 4
char_count = count_character('mississippi', 's')
print(char_count)

print(repeat_strings('hello', 3))

name1 = input('Please enter a FIRST NAME followed by a LAST NAME separated by a space:')
name2 = input('Please enter another FIRST NAME followed by a LAST NAME separated by a space:')

new_first_name1 = new_name1(name1, name2)
new_last_name1 = new_last_name(name1, name2)

new_first_name2 = new_name1(name2, name1)
new_last_name2 = new_last_name(name2, name1)


print(new_first_name1, new_last_name1)
print(new_first_name2, new_last_name2)

```
